```python
import cudf
import os
import cupy as cp
```

Import three separate CSVs and drop (misnamed) duplicate columns water_temp_4.1 and salinity_4.1 which contains the same data from water_temp_4 and salinity_4

```python
DATA_PATH = '../../data/hycom'
RESULTS_PATH = '../../results/hycom'

csv_1_path = os.path.join(DATA_PATH, 'HYCOM1.csv')
csv_2_path = os.path.join(DATA_PATH, 'HYCOM2.csv')
csv_3_path = os.path.join(DATA_PATH, 'HYCOM3.csv')


df_1 = cudf.read_csv(csv_1_path).drop(['water_temp_4.1', 'salinity_4.1'],axis=1)
df_2 = cudf.read_csv(csv_2_path).drop(['water_temp_4.1', 'salinity_4.1'],axis=1)
df_3 = cudf.read_csv(csv_3_path).drop(['water_temp_4.1', 'salinity_4.1'],axis=1)
```

```python
df_1 = df_1.drop(index=61032)
df_1.head()
```

```python
float_cols = ['lat', 'lon', 'water_temp_0', 'salinity_0',
       'water_temp_2', 'salinity_2', 'water_temp_4', 'salinity_4',
       'water_temp_6', 'salinity_6', 'water_temp_8', 'salinity_8']

for col in float_cols:
    df_1[col] = cudf.to_numeric(df_1[col])
```

# Join Dataframes

```python
df = df_1.append(df_2, ignore_index=True)
df = df.append(df_3, ignore_index=True)

df.head()
```

# Convert Date to DateTime

```python
df.start_date = cudf.to_datetime(df.start_date)
df.start_date = df.start_date.dt.strftime("%Y-%m-%d")
```
